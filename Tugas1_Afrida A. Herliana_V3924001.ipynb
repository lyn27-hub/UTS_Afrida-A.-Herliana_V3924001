```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
pelanggan = 8  # Rata-rata kedatangan pelanggan per jam
jam_hari = 12
hari = 30

# Simulasi kedatangan pelanggan
total_jam = jam_hari * hari
jumlah_kedatangan = np.random.poisson(pelanggan, total_jam)

# Visualisasi histogram
plt.figure(figsize=(10, 6))
plt.hist(jumlah_kedatangan, bins=range(0, max(jumlah_kedatangan) + 1), alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Histogram Jumlah Kedatangan Pelanggan')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Frekuensi')
plt.xticks(range(0, max(jumlah_kedatangan) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()

# Menghitung probabilitas tidak ada pelanggan dalam 1 jam
prob_tidak_ada = poisson.pmf(0, pelanggan)

# Hasil probabilitas
print(f"Probabilitas tidak ada pelanggan dalam 1 jam: {prob_tidak_ada:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
pelanggan = 8  # Rata-rata kedatangan pelanggan per jam
jam_hari = 12
hari = 30

# Simulasi kedatangan pelanggan
total_jam = jam_hari * hari
jumlah_kedatangan = np.random.poisson(pelanggan, total_jam)

# Visualisasi histogram
plt.figure(figsize=(10, 6))
plt.hist(jumlah_kedatangan, bins=range(0, max(jumlah_kedatangan) + 1), alpha=0.5, color='yellow', edgecolor='black', density=True, label='Simulasi')  # Histogram normalisasi

# Distribusi teoritis Poisson
x = np.arange(0, max(jumlah_kedatangan) + 1)
pmf = poisson.pmf(x, pelanggan)  # Fungsi massa probabilitas Poisson
plt.plot(x, pmf, 'o-', color='blue', label='Teoritis Poisson')  # Kurva distribusi Poisson

# Menambahkan label dan judul
plt.title('Perbandingan Distribusi Kedatangan Pelanggan: Simulasi vs Teoritis', fontsize=16)
plt.xlabel('Jumlah Pelanggan', fontsize=14)
plt.ylabel('Frekuensi Relatif', fontsize=14)
plt.xticks(range(0, max(jumlah_kedatangan) + 1))
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()

# Menghitung probabilitas tidak ada pelanggan dalam 1 jam
prob_tidak_ada = poisson.pmf(0, pelanggan)

# Menghitung probabilitas lebih dari 10 pelanggan dalam satu jam
prob_lebih_dari_10 = 1 - poisson.cdf(10, pelanggan)

# Hasil probabilitas
print(f"Probabilitas tidak ada pelanggan dalam 1 jam: {prob_tidak_ada:.4f}")
print(f"Probabilitas lebih dari 10 pelanggan dalam satu jam: {prob_lebih_dari_10:.4f}")
```
